## Data cleaning

Now we've looked around a bit at the data, we're going to clean it. These type of commands in Stata are:
* ``rename``
* ``replace``
* ``gen``
* ``collapse``
* ``merge``

First let's get these column names in English...

```python
# I'm going to reassign all variables to english names
sinac.columns
```

```python
sinac.columns = ['mother_age','no_pregnancies','length','weight',
             'mun_res', 'state_res','mun_birth','state_birth',
             'no_appts','schooling','day_birth', 'mo_birth','yr_birth',
             'sex','gest_age','apgar','type','type_doctor']
sinac.columns
```

```python
# wait... I don't like a name. Here's how to rename just one column:
sinac.rename({'type':'procedure'}, axis=1, inplace = True) # (rename)
```

```python
sinac.head()
```

```python
# Let's change the state code from 99 to missing (replace)
sinac.loc[sinac.state_res == 99,:] = np.NaN
sinac.state_res.describe() # good; the max is now 32
```

```python
# (gen)
# I want a numeric variable for infact sex =1 for female
sinac['sex_numeric'] = 0
sinac.loc[sinac.sex == 'Female','sex_numeric'] = 1

## OR, use numpy's "where"; where sex = Female, set sex_num = 1. 
sinac['sex_num'] = np.where((sinac['sex']  == 'Female'),1,0)

## let's use where for multiple categories
sinac['schooling_category'] = np.where((sinac.schooling < 4), 
                                       1,
                                       np.where((sinac.schooling >= 4) & (sinac.schooling <= 6),
                                               2,
                                               3))
```

```python
sinac.head()
```

```python
## this is equivalent to stata's:
# collapse (mean) [varlist], by(state_res mun_res)
sinac.groupby(['state_res','mun_res']).agg('mean').reset_index()
```

```python
# collapse (mean) [varlist] (std) [varlist], by(state_res mun_res)
sinac.groupby(['state_res','mun_res']).agg(['mean','std']).reset_index()
```

```python
# subindexing this:
sinac.groupby(['state_res','mun_res']).agg(['mean','std']).reset_index()['mother_age']['mean']
```

### Now let's pull in the dictionary file above. 
Say I want to add to this dataset the population of each municipality. First I need to make sure my columns match, then I can merge. Also, my dictionary is at the locality level, but I need the municipal level. So I'm going to follow these steps:
1. Collapse dictionary to municipal level
2. Rename columns
3. Merge

To merge, we will use the ``merge`` command. This takes two datasets, merges them on an array-type "on" option, and you can choose the following "how":
* right: This is equivalent to keeping only _merge == 2 | _merge == 3 
* left: This is equivalente to keeping only _merge == 1 | _merge == 3
* inner: Keep only _merge == 3
* outer: Keep all 

Note that the default for merge is "inner", but this may not be ideal for you!! I recommend always being explicit about what option you want. 

There is another option in merge that is really nice for those of us used to stata:
the "Indicator" option has a default of False, but if you set it to "True", it will add a variable to the dataset that takes on the _merge values of stata: "right_only", "left_only", "both", like _m = 1, _m =2, or _m = 3. 

```python
local.head()
```

```python
# 1. Get population totals by municipality, state
local_mun = local.groupby(['CVE_ENT','CVE_MUN']).POB_TOTAL.agg('sum').reset_index()
```

```python
local_mun.head()
```

```python
# 2. rename columns
local_mun.columns = ['state_res','mun_res','total_pop']
```

```python
# 3. merge
test = sinac.merge(local_mun, on = ['state_res','mun_res'],how = 'outer', indicator = True)
```

```python
# Let's look at our indicatory variable:
test[test._merge == 'left_only'].mun_res.value_counts()
# looks like all municipalities that didn't make the merge are 999, probably missing 
```

```python
# We don't care about municipalities that aren't in our data, so let's just keep
# matched and left-only:
sinac_merge = sinac.merge(local_mun, on = ['state_res','mun_res'],
                          how = 'left', indicator = False)
```
